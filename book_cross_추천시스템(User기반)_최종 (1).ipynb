```python
import numpy as np
import pandas as pd
import random
import warnings
warnings.filterwarnings("ignore")
from google.colab import drive

# 데이터 로드

drive.mount('/gdrive', force_remount=True)
df = pd.read_csv('/gdrive/My Drive/temp/book_df.csv', index_col='Unnamed: 0')
# df = pd.read_csv(cur_dir + '/data/book_cross_data.csv)
```

```python
df['book_author'].replace('Not Applicable (Na )', np.nan, inplace = True)
df['Category'].replace('9', np.nan, inplace=True)
df.dropna(inplace = True)
```

```python
df.info()
```

```python
# 책제목 입력 모델(유저-평점 기반 추천)
def get_best_rating_userid_title(title, n):
  best_rating_user = df[df['book_title'] == title].sort_values('rating', ascending=False)['user_id'][:n].values
  return best_rating_user

def best_rating_user_book_title(user_id):
  for id in user_id:
    recom_books = pd.DataFrame()
    df_combined = df[df['user_id'] == id].sort_values('rating', ascending=False)
    recom_books = pd.concat([recom_books, df_combined])
    recom_books = recom_books.sort_values(by='rating', ascending=False).drop_duplicates(subset = 'book_title', keep = 'first')
  return recom_books['book_title'][:10]

def recom_user_result_title(title):
  user_id = get_best_rating_userid_title(title, 5)
  books = best_rating_user_book_title(user_id)
  books = books[books != title]
  return books


# 작가 입력 모델(유저-평점 기반 추천)
def get_best_rating_userid_author(author, n):
  best_rating_user = df[df['book_author'] == author].sort_values('rating', ascending=False)['user_id'][:n].values
  return best_rating_user

def best_rating_user_book_author(user_id):
  for id in user_id:
    recom_books = pd.DataFrame()
    df_combined = df[df['user_id'] == id].sort_values('rating', ascending=False)
    recom_books = pd.concat([recom_books, df_combined])
    recom_books = recom_books.sort_values(by='rating', ascending=False).drop_duplicates(subset = 'book_title', keep = 'first')
  return recom_books['book_title'][:10]

def recom_user_result_author(author):
  user_id = get_best_rating_userid_author(author, 5)
  books = best_rating_user_book_author(user_id)
  books = books[books != author]
  return books
```

```python
title =input('제목 입력: ')
print('입력한 책에 대한 다른 사용자들의 추천 책은 다음과 같습니다.')
recom_book_by_title = recom_user_result_title(title)
recom_book_by_title = recom_book_by_title.tolist()
recom_book_by_title
```

```python
author =input('작가 입력: ')
print('입력한 작가에 대한 다른 사용자들의 추천 책은 다음과 같습니다.')
recom_book_by_author = recom_user_result_author(author)
recom_book_by_author = recom_book_by_author.tolist()
recom_book_by_author
```
